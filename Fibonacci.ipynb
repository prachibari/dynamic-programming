# DYBAMIC PROGRAMMING

A way of making your algorithm more efficient by storing some of the intermediate results. Works very well when your code has some repititive computation so you dont have to repeat those computations over and over again.

### Steps
1. Recursion
2. Store (Memoize)
3. Bottom Up

```python
### Finding the nth fibonacci series number

def fib(n):
    if n==1 or n==2:
        result = 1
    else:
        result = fib(n-1)+fib(n-2)
    return result
print(fib(4))
```

We can see that we are calculating many values more than once ie fib(20), fib(3) etc.
Say if we wanted to calculate fib(100), this is going to include redundant computations increasing the computation time.
#### Execution time
T(2^n)

#### Using Dynamic Programming
Store these values which are going to used repeatedly say fib(3),fib(2) etc

### Using Memoization Solution
1. Create an array of size n+1 and store the fib(n) at the corresponding index

```python
import time
memo={}
def fib(n):
    if n in memo: # we already have computed this fib
        return memo[n]
    # Note the below code is executed atmost n times
    if n==1 or n==2:
        result= 1
    else:
        result = fib(n-1)+fib(n-2) #fib is called twice
    
    #cache the value and return it
    memo[n] = result
    return result
print(fib(4))
```

#### Time complexity of above code:
If and else statement take O(1) time
### t < (2n+1).O(1) = O(n)

```python
from functools import lru_cache

@lru_cache(maxsize=1000)
def fib(n):
    if type(n) != int:
        raise TypeError("n must be positive integer")
    if n < 1:
        raise TypeError("n must be positive integer")
    if n==1 or n==2:
        result = 1
    else:
        result = fib(n-1)+fib(n-2)
    return result
print(fib(6))
```
